```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
london = pd.read_csv('london_ohe.csv')
london
```

```python
crime_columns = []

for column in london.columns:
    if 'crime' in column:
        crime_columns.append(column)
        
crime_columns


crime_count = london.groupby('month')[crime_columns].sum()
crime_count        
```

```python
stringency = london[['month', 'stringency']].drop_duplicates()
stringency
```

```python
crimes_count_stringency = pd.merge(crime_count, stringency, on="month")
crimes_count_stringency
```

```python
crimes_count_stringency.describe()
```

```python
cols = crimes_count_stringency.corr().nlargest(14, 'stringency')['stringency'].index
correlation_matrix = crimes_count_stringency[cols].corr()
```

```python
plt.figure(figsize = (15,10))
#correlation_matrix = crimes_count_stringency.corr() 

sns.heatmap(data = correlation_matrix, annot = True)
```
